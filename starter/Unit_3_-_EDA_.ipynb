#### Fill in blanks with a brief description of each plot type:

  
- Histogram
> A histogram groups values into ranges (or bins) and the height of a bar shows how many values fall in that range.It's best to use this type of plot with a variable that is made up of continuous numeric values.
- Density Plot / KDE
> The kernel density estimate plot creates a smooth version of a histogram by normalizing all points to appear under one curve. It's best to use this type of plot when you want to get a better idea of the shape of the distrubution of a dataset since it wouldn't be influenced by the number of bins used in it's corresponding histogram.
- Bar Chart
> A bar chart represents data with rectangular bars at heights or lengths proportional to the values they represent. It's best to use this type of chart when you want to show a distribution of data points or perform a comparison of metric values across different subgroups of your data.
- Line Plot
> A line plot connects the data points in a dataset via line segments. It's best to use this type of plot when you want to show a change in a variable over time.
- Univariate scatter plot
> A scatter plot uses dots to represent data points in a dataset. It's best to use this type of plot to showing a relationship betwen two numeric variables.
- Violin Plot
> A violin plot is a combination of a box plot and density plot that shows the distribution shape of the data. You can use this type of plot to view the distrubution of the data instead of a histogram and get the statistical summary as well.
- Box Plot
> A boxplot shows the distribution, center and skewness of a dataset. It's best to use this plot when you want to visually show the statistical summary of numeric values of a dataset.  
- Pie Chart
> A pie chart displays the percentage distribution of a categorical variable in a circular graph. This is best used when yo want to show a part-to-whole relationship.
- Rug Plot
> A rug plot draws ticks on the x-axis that show the location of individual data points. A rug plot can be helpful complement to a kde plot.

### Step 3 - Use Pearsonr & Spearmanr To Answer Questions About Correlation and P-Values

<h3 style='color:red;'> First, load 'salary_numeric.csv' - where education has been re-expressed as numeric variables.</h3>

```python
# load salary_numeric.csv into dataframe where education levels have been re-expressed as numeric
salary_numeric_df=pd.read_csv('salary_numeric.csv')
```

```python
################################### TO DO #########################################
####### FOLLOW PROJECT INSTRUCTIONS TO COMPLETE THIS STEP IN THE CELLS BELOW ######
```

```python
# import pearson and spearmanr methods from scipy.stats library
from scipy.stats import pearsonr, spearmanr
```

### What is the correlation between salary and education?

#### Your Hypothesis = Salary of survery respondents increase with the level of education obtained.

```python
# calculate correlation and pvalue of salary and education with spearmanr method
spearmanr(salary_numeric_df.salary, salary_numeric_df.education)
```

There is a positive correlation between salary and education.

#### Are these results statistically significant? How do you know?

The results are statistically significant because the p-value is below the thresold of .05.

### What is the correlation between salary and age?

#### Your Hypothesis = Salary of survery respondents increase with age.

```python
# calculate correlation and pvalue of salary and age with pearsonr method
pearsonr(salary_numeric_df.salary, salary_numeric_df.age)
```

There is a positive correlation between salary and age.

#### Are these results statistically significant? How do you know? 

The results are statistically significant because the p-value is below the thresold of .05.

### What is the correlation between salary and industry experience? 

#### Your Hypothesis =  Salary of survery respondents increase with years of experience.

```python
# calculate correlation and pvalue of salary and industry experience with pearsonr method
pearsonr(salary_numeric_df.salary, salary_numeric_df.industry_experience)
```

There is a positive correlation between salary and industry experience.

#### Are these results statistically significant? How do you know?

The results are statistically significant because the p-value is below the thresold of .05.
